```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('TopGoalScorer.csv')
data.head()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
groups = data.groupby('Player')
groups
```

```python
data['Goals'] = [int(item.split()[0]) for item in data['Goals']]
data['Appearances'] = [int(item.split()[0]) for item in data['Appearances']]
data
```

```python
groups['Goals'].count().sort_values(ascending=False)
```

```python
golden_boots = groups['Goals'].count().sort_values(ascending=False)
players = golden_boots.keys()
boots = golden_boots

plt.figure(figsize=(10, 10))
plt.barh(players, boots, color='blue')

plt.xlabel('Footballers')
plt.ylabel('Total golden boots')
plt.title('1992-2021 Golden Boot Stats')
```

```python
club_groups = data.groupby('Club')
gb_clubs = club_groups['Goals'].count().sort_values(ascending=False)

plt.figure(figsize=(10, 10))
plt.barh(gb_clubs.keys(), gb_clubs, color='red')

plt.xlabel('Clubs')
plt.ylabel('Total golden boots')
plt.title('1992-2021 Golden Boot Stats Clubs')
```
